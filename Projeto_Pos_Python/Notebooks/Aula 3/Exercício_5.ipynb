```python
def f(num, mult, tempo):
    print('{} - Start of execution'.format(threading.current_thread().name))
    while num > -1:
        print('{}: {} x {} = {}'.format(threading.current_thread().name, num, mult, (num*mult)))
        time.sleep(tempo)
        num -= 1
    print('{} - End of execution'.format(threading.current_thread().name))
```

```python
t1 = threading.Thread(target=f, args=(10, 1, 1), name='Tab-1')
t2 = threading.Thread(target=f, args=(10, 2, 1), name='Tab-2')

t1.start()
t1.join()
t2.start()
```
