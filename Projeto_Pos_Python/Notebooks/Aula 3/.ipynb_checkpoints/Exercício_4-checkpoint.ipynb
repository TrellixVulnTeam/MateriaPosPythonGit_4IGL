```python
import csv
import pandas as pd
```

```python
def read_csv(file):
    df = pd.read_csv(file, sep=';', header=0)
    return df.values.tolist()
```

```python
class Customer:
    def __init__(self,lista):
        self.id = lista[0]
        self.name = lista[1]
        self.gender = lista[2]
        self.birthday = lista[3]
        self.email = lista[4]
        self.active = lista[5]
        
    def __str__(self):
        return (str(self.id),str(self.name))
```

```python
lista_customer = read_csv('customers.csv')
```
