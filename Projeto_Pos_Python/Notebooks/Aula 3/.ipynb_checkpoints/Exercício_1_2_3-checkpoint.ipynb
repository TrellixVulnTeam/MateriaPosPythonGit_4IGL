## Exerc√≠cio 1:

```python
# Utilizando Fun√ß√£o:
def F(n):
    if n == 1:
        return 2
    elif n == 2:
        return 1
    else:
        return (2 * F(n-1) + G(n-2))

def G(n):
    if n == 1:
        return 1
    elif n == 2:
        return 2
    else:
        return (G(n-1) + 3 * F(n-2))
    
def K(n):
    if n > 0:
        return F(n), G(n)
    else:
        return None
```

```python
print(K(2))
print(K(3))
print(K(4))
```

```python
# Utilizando lambda:
Fl = lambda x : 1 if x == 2 else 2 if x == 1 else (2 * Fl(x-1) + Gl(x-2))
Gl = lambda x : x if x < 3 else 2 if x == 1 else (Gl(x -1) + 3 * Fl(x-2))
Kl = lambda x : None if x == 0 else (Fl(x),Gl(x))
```

```python
print(Kl(2))
print(Kl(3))
print(Kl(4))
```

```python
# Utilizando 'List-Comprehension':
[K(i) for i in range(2,5)]
```

```python
# Utilizando 'Generator Expression':
result = (K(i) for i in range(2,5))

for i in result:
    print(i)
```

```python
from random import randint
```

```python
def tri_retangulo(x,y,z):
    if x > y and x > z:
        return True if x**2 == (y**2 + z**2) else False
    elif y > x and y > z:
        return True if y**2 == (x**2 + z**2) else False
    else:
        return True if z**2 == (x**2 + y**2) else False
```

```python
tri_retangulo(13,5,12)
```

```python
tri_retangulo(12,5,13)
```

```python
tri_retangulo(12,13,5)
```

```python
tri_retangulo(5,13,5)
```
